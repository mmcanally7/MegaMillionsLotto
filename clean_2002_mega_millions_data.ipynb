```python
df.rename(columns={"Draw Date": "date", "Mega Ball": "mega_ball", "Multiplier": "mega_plier"}, inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
ball_1 = np.zeros(df.shape[0], int)
ball_2 = np.zeros(df.shape[0], int)
ball_3 = np.zeros(df.shape[0], int)
ball_4 = np.zeros(df.shape[0], int)
ball_5 = np.zeros(df.shape[0], int)

print(len(ball_1))
```

```python
for i, row in enumerate(df['Winning Numbers']):
    ball_1[i]=row[0:2]
    ball_2[i]=row[3:5]
    ball_3[i]=row[6:8]
    ball_4[i]=row[9:11]
    ball_5[i]=row[12:14]
    
print(ball_1)    
```

```python
df['ball_1']=ball_1
df['ball_2']=ball_2
df['ball_3']=ball_3
df['ball_4']=ball_4
df['ball_5']=ball_5
```

```python
df.head()
```

```python
df.tail()
```

```python
df = df.drop('Winning Numbers', axis=1)
df.head()
```

```python
cols = df.columns.tolist()
cols
```

```python
cols=cols[0:-7]+cols[-5:]+cols[1:-5]
cols
```

```python
df=df[cols]
```

```python
df.head()
```

```python
df.tail()
```

```python
df['date']=pd.to_datetime(df['date'])
df['ball_1']=df['ball_1'].astype(float)
df['ball_2']=df['ball_2'].astype(float)
df['ball_3']=df['ball_3'].astype(float)
df['ball_4']=df['ball_4'].astype(float)
df['ball_5']=df['ball_5'].astype(float)
df['mega_ball']=df['mega_ball'].astype(float)
df['mega_plier']=df['mega_plier'].astype(float)
```

```python
type(df['date'])
```

```python
df.sort_values(by='date', inplace = True, ascending=False) 
```

```python
df['date']=df['date'].dt.strftime("%m/%d/%Y")
```

```python
df.head()
```

```python
df=df.fillna(0)
df.head()
```
