```python
reg1_url = "https://www.megamillions.com/Winning-Numbers/Previous-Drawings.aspx?pageNumber=1&pageSize=20&startDate=01/01/2017&endDate=01/01/2020#page19"
driver.get(reg1_url)

html1 = driver.page_source
soup1=BeautifulSoup(html1,'lxml')
```

```python
reg2_url = "https://www.megamillions.com/Winning-Numbers/Previous-Drawings.aspx?pageNumber=1&pageSize=20&startDate=01/01/2014&endDate=01/01/2017#page19"
driver.get(reg2_url)

html2 = driver.page_source
soup2=BeautifulSoup(html2,'lxml')
```

```python
reg3_url = "https://www.megamillions.com/Winning-Numbers/Previous-Drawings.aspx?pageNumber=1&pageSize=20&startDate=01/01/2011&endDate=01/01/2014#page19"
driver.get(reg3_url)

html3 = driver.page_source
soup3=BeautifulSoup(html3,'lxml')
```

```python
reg4_url = "https://www.megamillions.com/Winning-Numbers/Previous-Drawings.aspx?pageNumber=1&pageSize=20&startDate=01/01/2010&endDate=01/01/2011#page5"
driver.get(reg4_url)

html4 = driver.page_source
soup4=BeautifulSoup(html4,'lxml')
```

```python
print(soup3.prettify)
```

```python
ball_1 = []
ball_2 = []
ball_3 = []
ball_4 = []
ball_5 = []
mega = []
mega_plier = []
dates = []

data_sets = [ball_1, ball_2, ball_3, ball_4, ball_5, mega, mega_plier, dates]

soups = [soup, soup1, soup2, soup3, soup4]

for sp in soups:
    all_li = sp.find_all('li')
    for line in all_li:
        if line.get('class') == ['ball', 'pastNum1']:
            ball_1.append(line.string)
        elif line.get('class') == ['ball', 'pastNum2']:
            ball_2.append(line.string)    
        elif line.get('class') == ['ball', 'pastNum3']:
            ball_3.append(line.string)       
        elif line.get('class') == ['ball', 'pastNum4']:
            ball_4.append(line.string)   
        elif line.get('class') == ['ball', 'pastNum5']:
            ball_5.append(line.string) 
        elif line.get('class') == ['ball', 'yellowBall', 'pastNumMB']:
            mega.append(line.string)  

for data in data_sets:
    print(len(data))   
```

```python
for sp in soups:
    all_h5=sp.find_all('h5')
    for line in all_h5:
        if line.get('class') == ['ball', 'pastNum1']:
            ball_1.append(line.string)
        elif line.get('class') == ['ball', 'pastNum2']:
            ball_2.append(line.string)    
        elif line.get('class') == ['ball', 'pastNum3']:
            ball_3.append(line.string)       
        elif line.get('class') == ['ball', 'pastNum4']:
            ball_4.append(line.string)   
        elif line.get('class') == ['ball', 'pastNum5']:
            ball_5.append(line.string) 
        elif line.get('class') == ['ball', 'yellowBall', 'pastNumMB']:
            mega.append(line.string) 
        elif line.get("class") == ['drawItemDate']:
            dates.append(line.string)

for data in data_sets:
    print(len(data))       
    
print(dates)  
```

```python
for sp in soups:
    all_span=sp.find_all('span')
    for line in all_span:
        if line.get('class') == ['megaplier', 'pastNumMP']:
            mega_plier.append(line.string)
        
for data in data_sets:
    print(len(data))  
```

```python
from functools import reduce
import re
 
def remove_char(lst, char):
    return reduce(lambda x, y: x + [re.sub(char, '', y)], lst, [])

char='X'

mega_pl = remove_char(mega_plier,char)
```

```python
print(mega_pl[0:10])
```

```python
ball_1 = [float(str(elem)) for elem in ball_1]
ball_2 = [float(str(elem)) for elem in ball_2]
ball_3 = [float(str(elem)) for elem in ball_3]
ball_4 = [float(str(elem)) for elem in ball_4]
ball_5 = [float(str(elem)) for elem in ball_5]
mega = [float(str(elem)) for elem in mega]
mega_pl = [float(str(elem)) for elem in mega_pl]
```

```python
mega_mil={'date':dates, 'ball_1':ball_1, 'ball_2':ball_2, 'ball_3':ball_3, 'ball_4':ball_4, 'ball_5':ball_5, 'mega_ball':mega, 'mega_plier':mega_pl}
```
